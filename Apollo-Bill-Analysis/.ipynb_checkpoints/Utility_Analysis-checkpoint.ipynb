```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### First Look

```python
df = pd.read_csv("Consumers_2022_2024.csv")
df.head()
```

### Now a Basic Usage Graph

```python
# Assuming your data has a column 'BILL_MONTH_YEAR'
df["Date"] = pd.to_datetime(df.BILL_MONTH_YEAR, format='%b-%y')

# Define summer months (replace with your definition if needed)
summer_months = set(['Jun', 'Jul', 'Aug'])

# Extract year information from 'Date'
df['Year'] = df['Date'].dt.year

# Calculate start and end dates for summer each year
summer_start = pd.to_datetime([f'{year}-06-01' for year in df['Year'].unique()])
summer_end = pd.to_datetime([f'{year}-08-31' for year in df['Year'].unique()])

# Plot the data with markers
plt.plot(df.Date, df.TOTAL_USAGE_COL, marker='o', linestyle='-');

# Plot vertical lines for summer start and end (using 'axvline')
for start, end in zip(summer_start, summer_end):
  plt.axvline(x=start, color='r', linestyle='--')
  plt.axvline(x=end, color='r', linestyle='--')

# Customize the plot
plt.xlabel("Date")
plt.ylabel("Usage (McF)")
plt.xticks(rotation=45)
plt.legend()
plt.title("Consumer Usage, 2022-2025")

plt.show()
```

**From here it's pretty clear that there is a clean dropoff for the summer months. This makes sense between the gas system is automatic. From here, a logical goal for improving sustainability and costs is to simply reduce usage.**
